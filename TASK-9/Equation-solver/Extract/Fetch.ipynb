# Building the CNN model

```python
model = Sequential()
model.add(Conv2D(35, (3,3), input_shape=(28, 28,1), activation='relu',padding='same'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Conv2D(14, (3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.2))
model.add(Flatten())
model.add(Dense(129, activation='relu'))
model.add(Dense(50, activation='relu'))
model.add(Dense(13, activation='softmax'))
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
from keras.models import Sequential
from keras.layers import Dense

# Define the model architecture
model = Sequential()
model.add(Dense(128, input_shape=(X_r.shape[1],), activation='relu'))
model.add(Dense(13, activation='softmax'))

# Compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Now, you can train the model with your data (X_r and lab)
model.fit(X_r, lab, epochs=11, batch_size=200, shuffle=True, verbose=1)

# Save the weights after training
model.save_weights("model_final.h5")
```
